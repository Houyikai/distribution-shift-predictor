```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

mpl.rcParams['font.size'] = 16
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Times New Roman'

dataset = 'ETTh1'

df1 = pd.read_csv(f'./data/sparam_sensitive_0_{dataset}.csv',index_col='Unnamed: 0')
df1 = df1.T

df2 = pd.read_csv(f'./data/sparam_sensitive_1_{dataset}.csv',index_col='Unnamed: 0')
df2 = df2.T

df3 = pd.read_csv(f'./data/sparam_sensitive_2_{dataset}.csv',index_col='Unnamed: 0')
df3 = df3.T

df4 = pd.read_csv(f'./data/sparam_sensitive_3_{dataset}.csv',index_col='Unnamed: 0')
df4 = df4.T

# 创建一个2x2的图形网格
fig, axs = plt.subplots(1, 4, figsize=(20, 3),dpi=300)
plt.subplots_adjust(wspace=0.4)

# ax0
x_ticks = [1,7,17,25,49]
axs[0].set_xticks(x_ticks)
axs[0].grid(True)
axs[0].set_ylabel('MSE')
axs[0].set_xlabel('MNK size(\u03C9)')

axs[0].plot(x_ticks,df1[48],label='H = 48',marker='o')
axs[0].plot(x_ticks,df1[96],label='H = 96',marker='*')
axs[0].plot(x_ticks,df1[192],label='H = 192',marker='.')
axs[0].plot(x_ticks,df1[336],label='H = 336',marker=',')
axs[0].plot(x_ticks,df1[720],label='H = 720',marker='x')

# ax1
x_ticks = [1,2,3,4,5]
axs[1].set_xticks(x_ticks)
axs[1].grid(True)
axs[1].set_ylabel('MSE')
axs[1].set_xlabel('DSP layers(\u2113)')

axs[1].plot(x_ticks,df2[48],label='H = 48',marker='o')
axs[1].plot(x_ticks,df2[96],label='H = 96',marker='*')
axs[1].plot(x_ticks,df2[192],label='H = 192',marker='.')
axs[1].plot(x_ticks,df2[336],label='H = 336',marker=',')
axs[1].plot(x_ticks,df2[720],label='H = 720',marker='x')

# ax2
x_ticks = [48, 96, 128, 192, 336]
axs[2].set_xticks(x_ticks)
axs[2].grid(True)
axs[2].set_ylabel('MSE')
axs[2].set_xlabel('Attn receive length(\u03B6)')

axs[2].plot(x_ticks,df3[48],label='H = 48',marker='o')
axs[2].plot(x_ticks,df3[96],label='H = 96',marker='*')
axs[2].plot(x_ticks,df3[192],label='H = 192',marker='.')
axs[2].plot(x_ticks,df3[336],label='H = 336',marker=',')
axs[2].plot(x_ticks,df3[720],label='H = 720',marker='x')

# ax3
x_ticks = [192, 336, 512, 720, 1000]
axs[3].set_xticks(x_ticks)
axs[3].grid(True)
axs[3].set_ylabel('MSE')
axs[3].set_xlabel('Input length(L)')

axs[3].plot(x_ticks,df4[48],label='H = 48',marker='o')
axs[3].plot(x_ticks,df4[96],label='H = 96',marker='*')
axs[3].plot(x_ticks,df4[192],label='H = 192',marker='.')
axs[3].plot(x_ticks,df4[336],label='H = 336',marker=',')
axs[3].plot(x_ticks,df4[720],label='H = 720',marker='x')


# 创建一个图例并将其放置在图形的右上角
lines, labels = fig.axes[-1].get_legend_handles_labels()
fig.legend(lines, labels, loc='lower center', ncol=5,bbox_to_anchor=(0.5, -0.3))
fig.text(0.5, -0.5, dataset, ha='center', va='bottom', fontsize=22)
plt.show()
```
