```python

import pandas as pd
dataset = 'Weather'

dsp = pd.read_csv(f'./data/{dataset}_336_dsp.csv')
tst = pd.read_csv(f'./data/{dataset}_336_tst.csv')
result = pd.merge(left=tst,right=dsp,how='outer')
from matplotlib import pyplot as plt
import matplotlib as mpl
plt.figure(figsize=(20,5),dpi=300)
# plt.grid(True)
mpl.rcParams['font.size'] = 20
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Times New Roman'

startpoint = 0
length = 720
plt.plot(result[startpoint:startpoint+length].y.values, label='Actual')
plt.plot(result[startpoint:startpoint+length].PatchTST.values, label='PatchTST')
plt.plot(result[startpoint:startpoint+length].MIXModel.values, label='Ours method',color='#008B45')
plt.xlabel('time')
plt.ylabel('value')
plt.legend() 
```

```python
temp = result[result['unique_id']=='wd (deg)']
plt.figure(figsize=(20,5),dpi=300)
# plt.grid(True)
mpl.rcParams['font.size'] = 20
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Times New Roman'

startpoint = 0
length = 720
plt.plot(temp[startpoint:startpoint+length].y.values, label='Actual')
plt.plot(temp[startpoint:startpoint+length].PatchTST.values, label='PatchTST')
plt.plot(temp[startpoint:startpoint+length].MIXModel.values, label='Ours method',color='#008B45')
plt.xlabel('time')
plt.ylabel('value')
plt.legend(loc='lower right')
```
