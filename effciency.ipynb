```python
import pandas as pd
import numpy as np
df = pd.read_csv('./exp_results/time.csv',index_col='Unnamed: 0')
df['mixmodel'][96] = 4.252171182632446
df = np.log10(df)
time_count = df
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.rcParams['lines.linewidth'] = 1
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = ['Times New Roman']
mpl.rcParams['font.size'] = 20
fig = plt.figure(figsize=(4, 3),dpi=300)
plt.grid(True, linewidth=0.6, color='gray', linestyle='-')
x_ticks = [96,192,336,720]
plt.xticks(x_ticks)
plt.ylabel('Computational Time\n[log scale, Seconds]')
plt.xlabel('Horizon')
plt.plot(time_count["mixmodel"], label="Ours mothod",color = '#008B45', marker='o')
plt.plot(time_count["patchtst"], label="PatchTST", marker='v')
plt.plot(time_count["nhits"], label="NHITS" ,marker='*')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3),frameon=False)
fig.text(0.5, -0.8, '(a) Time Efficiency', ha='center', va='bottom', fontsize=22)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
params_count = pd.read_csv('./exp_results/memory.csv',index_col=['Unnamed: 0'])
import numpy as np
plt.rcParams['lines.linewidth'] = 1
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = ['Times New Roman']
mpl.rcParams['font.size'] = 20
fig = plt.figure(figsize=(4, 3), dpi=300)
plt.grid(True, linewidth=0.6, color='gray', linestyle='-')

y_ticks = np.linspace(0,14000000,8)
y_ticks_b = np.arange(0,15,2)
plt.yticks(y_ticks,y_ticks_b)
plt.ylabel('Number of Parameters\n[Millions]')
plt.xlabel('Horizon')
plt.plot(params_count["mixmodel"], label="Ours mothod",color = '#008B45', marker='o')
plt.plot(params_count["patchtst"], label="PatchTST", marker='v')
plt.plot(params_count["nhits"], label="NHITS" ,marker='*')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3),frameon=False)
fig.text(0.5, -0.8, '(b) Memory Efficiency', ha='center', va='bottom', fontsize=22)
plt.show()
```
