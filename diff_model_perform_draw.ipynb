```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd


dataset = 'ETTh1'
horizon = 192
def draw(modelname):
  df = pd.read_csv(f'./data/{dataset}_{horizon}_{modelname}.csv')
  data = df[df['unique_id']=='LUFL'][['ds','y',modelname]][0:192]

  plt.figure(figsize=(4,3),dpi=300)
  mpl.rcParams['font.size'] = 12
  mpl.rcParams['font.family'] = 'serif'
  mpl.rcParams['font.serif'] = 'Times New Roman'
  plt.plot(data['y'])
  plt.plot(data[modelname])
  plt.suptitle(modelname, y=-0.05)
  plt.savefig(f'F:\文档\专利和论文参考\overleaf_\performance_{modelname}.png', bbox_inches='tight', pad_inches=0.1)
  print(f'{modelname} done')
# Autoformer, FEDformer, Informer, PatchTST, NBEATS, NHITS, MLP
modelnames = ['MIXModel','Autoformer','FEDformer','Informer','PatchTST','NBEATS','NHITS','MLP']
for modelname in modelnames:
  draw(modelname)
```
