```python
from neuralforecast.models import HEncoder
from neuralforecast.losses.pytorch import MAE
from datasetsforecast.long_horizon2 import LongHorizon2, LongHorizon2Info
from neuralforecast import NeuralForecast
import pandas as pd
import os

input_size = 512
horizon = 192
dataset = "ETTm1"

Y_df = LongHorizon2.load(directory="./data/", group=dataset)
Y_df["ds"] = pd.to_datetime(Y_df["ds"])
freq = LongHorizon2Info[dataset].freq
val_size = LongHorizon2Info[dataset].val_size
test_size = LongHorizon2Info[dataset].test_size

epochs = 100

hencoder = HEncoder(
    h=horizon,
    input_size=input_size,
    max_steps=epochs,
    exclude_insample_y=False,
    stat_exog_list=None,
    hist_exog_list=None,
    futr_exog_list=None,
    decoder_dropout = 0.0,
    decomp_kernel_size = 25,
    
    ### MLP:
    mlp_hidden_size = 512,
    mlp_stacks = 3,
    mlp_blocks = 1,  
    mlp_block_layers = 1, 
    n_pool_kernel_size = [16, 8, 1],
    n_freq_downsample=[168, 24, 1],  
    mlp_pooling_mode = "MaxPool1d",
    mlp_interpolation_mode = "linear",
    mlp_dropout=0.6,
    mlp_activation = "ReLU",

    ### former:
    former_input_size = 96,
    former_encoder_layers = 1,
    former_heads = 4,
    former_hidden_size = 32,
    former_linear_size = 512,
    former_dropout = 0.0,
    fc_dropout = 0.0,
    head_dropout = 0.0,
    attn_dropout = 0.6,
    former_patch_len = 24,
    former_stride = 12,
    revin = False,
    revin_affine = False,
    revin_subtract_last = True,
    former_activation = "gelu",
    res_attention = True,
    batch_normalization = True,
    learn_pos_embed = True,
    
    ### basics:
    loss=MAE(),
    valid_loss=None,
    learning_rate = 0.0027,
    num_lr_decays = -1,
    early_stop_patience_steps = 3,
    val_check_steps = 5,
    batch_size = 32,
    valid_batch_size = None,
    windows_batch_size=32,
    inference_windows_batch_size = 32,
    step_size = 1,
    scaler_type = None,
    random_seed = 1,
    num_workers_loader = 0,
    drop_last_loader = False,
)

nf = NeuralForecast(models=[hencoder], freq=freq)
nf.fit(df=Y_df, val_size=val_size)
```

```python
savepath:str = os.path.join('checkpoints',dataset+str(horizon))
nf.save(path=savepath,
        model_index=None, 
        overwrite=True,
        save_dataset=True)
```

```python
Y_hat_insample[-200:].plot(x='ds',y=['y','HEncoder'],figsize=(20,5))
```

```python
from neuralforecast.losses.numpy import mae, mse

Y_hat_df = Y_hat_insample
y_true = Y_hat_df.y.values
y_hat = Y_hat_df["HEncoder"].values
n_series = len(Y_hat_df.unique_id.unique())
y_true = y_true.reshape(n_series, -1, horizon)
y_hat = y_hat.reshape(n_series, -1, horizon)
print("\n" * 4)
print("Parsed results")
print(f"HEncoder {dataset} h={horizon}")
# print("test_size", test_size)
print("y_true.shape (n_series, n_windows, n_time_out):\t", y_true.shape)
print("y_hat.shape  (n_series, n_windows, n_time_out):\t", y_hat.shape)
# print(' best validation hyperparameter:\t', nf.models[0].results.get_best_result().config)
print("MSE: ", mse(y_hat, y_true))
print("MAE: ", mae(y_hat, y_true))
```
